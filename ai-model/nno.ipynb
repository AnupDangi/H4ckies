```python
import h2o
h2o.init()
```

```python

# Upload the dataset from a local file
data = h2o.import_file(r"C:\Users\Asifa Bandulal Beed\Downloads\drug_interactions_dataset.csv")
```

```python
# Set the target column
target = 'C6'

# Split data into training and testing sets
train, test = data.split_frame(ratios=[.8])
```

```python
from h2o.automl import H2OAutoML

# Initialize AutoML
aml = H2OAutoML(max_models=10, seed=1)

# Train the model
aml.train(y=target, training_frame=train)
```

```python
print(data.columns)
```

```python
# Evaluate the best model on the test data
performance = aml.leader.model_performance(test_data=test)
print(performance)
```

```python
# Make predictions on test data
predictions = aml.leader.predict(test)
```

```python
model_path=aml.leader.save_model(path="drug-interaction-aml-model")
print("model saved to :",model_path)
```

```python
# Save the best model from AutoML
model_path = h2o.save_model(model=aml.leader, path="drug_interaction_aml_model", force=True)
print("model saved to:",model_path)
```
