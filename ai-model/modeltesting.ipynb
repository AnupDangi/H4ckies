```python
import xml.etree.ElementTree as ET
import pandas as pd
import xmltodict

# Define a function to parse XML file and convert it to a DataFrame
def parse_xml_to_df(file_path, record_tag):
    # Open and parse the XML file
    with open(file_path, 'r') as file:
        xml_data = xmltodict.parse(file.read())
    
    # Convert the XML data into a list of dictionaries (records) for each item
    records = xml_data['root'][record_tag]  # Replace 'root' and 'record_tag' as per your XML structure

    # Convert list of dictionaries to a DataFrame
    return pd.DataFrame(records)

# Load FAERS data from XML files
drug_data = parse_xml_to_df("data/DRUG.xml", "DRUG")
reac_data = parse_xml_to_df("data/REAC.xml", "REAC")

# Filter for suspect drugs only
drug_data = drug_data[drug_data['role_cod'] == 'PS']

# Create drug pairs by CASE_ID
drug_pairs = drug_data.groupby('case_id')['drug_name'].apply(lambda x: list(x)).reset_index()

# Expand each pair into rows
drug_pairs_expanded = []
for _, row in drug_pairs.iterrows():
    drugs = row['drug_name']
    case_id = row['case_id']
    for i in range(len(drugs)):
        for j in range(i + 1, len(drugs)):
            drug_pairs_expanded.append((case_id, drugs[i], drugs[j]))

# Convert to DataFrame
drug_pairs_df = pd.DataFrame(drug_pairs_expanded, columns=['case_id', 'drug_a', 'drug_b'])

# Display result
print(drug_pairs_df.head())
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
from itertools import combinations

# Parse XML file
tree = ET.parse(r"C:\Users\Asifa Bandulal Beed\druginteraction_predictionml\XML\drugsh.xml")
root = tree.getroot()

# Lists to hold extracted data
case_ids = []
drug_names = []
role_codes = []

# Manually navigate through unstructured XML
for elem in root.iter():
    # Check for tags of interest and collect data accordingly
    if elem.tag == 'case_id':  # Adjust tag name based on your XML
        current_case_id = elem.text
    elif elem.tag == 'drug_name':  # Adjust tag name based on your XML
        drug_names.append(elem.text)
        case_ids.append(current_case_id)
    elif elem.tag == 'role_cod':  # Adjust tag name based on your XML
        role_codes.append(elem.text)

# Create DataFrame from the extracted lists
drug_data = pd.DataFrame({
    'case_id': case_ids,
    'drug_name': drug_names,
    'role_cod': role_codes
})

# Filter for suspect drugs only
drug_data = drug_data[drug_data['role_cod'] == 'PS']

# Create drug pairs by CASE_ID
drug_pairs = drug_data.groupby('case_id')['drug_name'].apply(lambda x: list(x)).reset_index()

# Expand each pair into rows
drug_pairs_expanded = []
for _, row in drug_pairs.iterrows():
    drugs = row['drug_name']
    case_id = row['case_id']
    
    # Skip cases with fewer than two drugs
    if len(drugs) < 2:
        continue
    
    # Generate all unique drug pairs using combinations
    pairs = list(combinations(drugs, 2))
    
    # Append each pair into the expanded list
    for pair in pairs:
        drug_pairs_expanded.append((case_id, pair[0], pair[1]))

# Convert to DataFrame
drug_pairs_df = pd.DataFrame(drug_pairs_expanded, columns=['case_id', 'drug_a', 'drug_b'])

# Display result
print(drug_pairs_df.head())

# Optionally, save to CSV
drug_pairs_df.to_csv('drug_pairs.csv', index=False)
```
